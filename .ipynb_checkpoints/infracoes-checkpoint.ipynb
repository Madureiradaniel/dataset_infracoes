```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_infracoes_abril = pd.read_csv('infracoes_abril.csv', sep=';')
df_infracoes_marco = pd.read_csv('infracoes_marco.csv', sep=';', low_memory=False)

#convertendo em data
df_infracoes_abril['cometimento'] = pd.to_datetime(df_infracoes_abril['cometimento'])
df_infracoes_marco['cometimento'] = pd.to_datetime(df_infracoes_abril['cometimento'])
```

```python
# gráfico de linha qtd de multas por periodo
ax = plt.gca()

df_infracoes_abril.groupby('cometimento')['tipo_infracao'].count().plot(kind='line', x='cometimento', y='tipo_infracao', ax=ax, color='red')
plt.show()
```
